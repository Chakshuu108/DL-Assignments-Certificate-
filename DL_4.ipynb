```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

link = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"

cols = ["sx","L","D","H","W","SW","VW","SH","R"]
df = pd.read_csv(link, names=cols)

df = pd.get_dummies(df, columns=["sx"])

X = df.drop("R", axis=1).values
Y = df["R"].values.reshape(-1,1)

sc = StandardScaler()
X = sc.fit_transform(X)

X_tr, X_te, Y_tr, Y_te = train_test_split(X, Y, test_size=0.2, random_state=0)

i_s = X_tr.shape[1]
h_s = 10
o_s = 1

A = np.random.randn(i_s, h_s)
a = np.zeros((1, h_s))
B = np.random.randn(h_s, o_s)
b = np.zeros((1, o_s))

lr = 0.001
ep = 1000

def rel(x):
    return np.maximum(0, x)

def rel_d(x):
    return (x > 0).astype(float)

for _ in range(ep):
    z1 = np.dot(X_tr, A) + a
    h1 = rel(z1)
    z2 = np.dot(h1, B) + b
    y_p = z2

    d2 = (y_p - Y_tr) / len(X_tr)
    dB = np.dot(h1.T, d2)
    db = np.sum(d2, axis=0, keepdims=True)

    d1 = np.dot(d2, B.T)
    dz1 = d1 * rel_d(z1)
    dA = np.dot(X_tr.T, dz1)
    da = np.sum(dz1, axis=0, keepdims=True)

    B -= lr * dB
    b -= lr * db
    A -= lr * dA
    a -= lr * da

t1 = np.dot(X_te, A) + a
t2 = rel(t1)
t3 = np.dot(t2, B) + b
mse = np.mean((t3 - Y_te)**2)

print("MSE:", mse)
```
