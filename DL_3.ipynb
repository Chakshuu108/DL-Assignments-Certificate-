```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

link = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
dt = pd.read_csv(link)

A = dt.iloc[:, :-1].values
B = dt.iloc[:, -1].values

scal = StandardScaler()
A = scal.fit_transform(A)

A_tr, A_te, B_tr, B_te = train_test_split(A, B, test_size=0.2, random_state=0)

w = np.zeros(A_tr.shape[1])
bi = 0
lrn = 0.001
epc = 1000

for _ in range(epc):
    out = np.dot(A_tr, w) + bi
    dw = (1/len(A_tr)) * np.dot(A_tr.T, (out - B_tr))
    db = (1/len(A_tr)) * np.sum(out - B_tr)
    w -= lrn * dw
    bi -= lrn * db

pred = np.dot(A_te, w) + bi

ms = np.mean((pred - B_te)**2)

print("MSE:", ms)
```
