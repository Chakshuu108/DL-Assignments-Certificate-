```python
def st(a):
    if a >= 0:
        return 1
    else:
        return 0

def pc(p, q, u, v, c):
    s = p*u + q*v + c
    return st(s)

inp = [(0,0), (0,1), (1,0), (1,1)]

print("AND Gate")
for p, q in inp:
    o = pc(p, q, 1, 1, -1.5)
    print(p, q, "->", o)

print()

print("OR Gate")
for p, q in inp:
    o = pc(p, q, 1, 1, -0.5)
    print(p, q, "->", o)

print()

print("NAND Gate")
for p, q in inp:
    o = pc(p, q, -1, -1, 1.5)
    print(p, q, "->", o)

print()

print("NOR Gate")
for p, q in inp:
    o = pc(p, q, -1, -1, 0.5)
    print(p, q, "->", o)
```
